```python
import csv
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
import joblib
```

```python
df = pd.read_csv('weight-height (1).csv')
df
```

```python
label_enc = LabelEncoder()
std_scaler = StandardScaler()

df.Gender = label_enc.fit_transform(df.Gender)
df[["Gender", "Height"]] = std_scaler.fit_transform(df[["Gender", "Height"]])
```

```python
df
```

```python
X = df[["Gender", "Height"]]
X
```

```python
y = df["Weight"]
y
```

```python
lin_regr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y)

model = lin_regr.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
```

```python
print(f"Akurasi Model: {accuracy * 100}%")
```

```python
joblib.dump((model, std_scaler), "weight-prediction-using-linear-regression.pkl")
```
